```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('train_data.csv')
train_data.head()
```

```python
train_data.info()
```

```python
train_data.columns
```

```python
train_data.isna().sum()
```

```python
train_data_cleaned = train_data.dropna()
```

```python
train_data_cleaned.isna().sum()
```

```python
train_data_cleaned.shape
```

```python
date_cols = ['date première tentative', 'date livraison/échec', 'date dernier statut']

for col in date_cols:
    train_data_cleaned[col+' duration'] = (pd.to_datetime(train_data_cleaned[col]) - pd.to_datetime(train_data_cleaned['date expédition'])).dt.total_seconds() / (60*60*42)
```

```python
train_data_cleaned.head()
```

```python
train_data_cleaned['stop desk'] = train_data_cleaned['stop desk?'].apply(lambda x: 0 if x=='oui' else 1)
```

```python
train_data_cleaned['stop desk'].value_counts()
```

```python
train_data_cleaned['destination_commune_wilaya_same'] = (train_data_cleaned['destination commune'] == train_data_cleaned['destination wilaya']).astype(int)
```

```python
train_data_cleaned['destination_commune_wilaya_same'].value_counts()
```

```python
train_data_cleaned = train_data_cleaned.drop(columns=['Tracking', 'date expédition', 'date première tentative', 'date livraison/échec', 'date dernier statut', 'stop desk?'])
```

```python
train_data_cleaned.head()
```

```python
train_data_cleaned.info()
```

```python
cat_features = ['ID expediteur', 'depart commune', 'depart wilaya', 'destination commune', 'destination wilaya']
```

```python
for c in cat_features:
    train_data_cleaned[c] = train_data_cleaned[c].astype('category')
```

```python
y = train_data_cleaned['dernier statut']
X = train_data_cleaned.drop(columns=['dernier statut', 'destination commune', 'depart commune'])
```

```python
X.head()
```

```python
y.head()
```

```python
d = dict(X['ID expediteur'].value_counts())
```

```python
X['ID expediteur'] = X['ID expediteur'].map(d)
```

```python
X.head()
```

```python
cat_columns = ['depart wilaya', 'destination wilaya']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
pipeline = Pipeline([
    ('encoder', OneHotEncoder())
])
```

```python
preprocessing = ColumnTransformer(
    transformers=[
        ('cat', pipeline, cat_columns)
    ],
    remainder='passthrough'
    )
```

```python
X_prepared = preprocessing.fit_transform(X)
```

```python
X_prepared = X_prepared.toarray()
```

```python
X_prepared.shape
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
import xgboost as xgb
```

```python
encoder = LabelEncoder()
```

```python
y_encoded = encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_prepared, y_encoded, test_size=0.2, random_state=42)
```

```python
xgb_clf = xgb.XGBClassifier()
xgb_clf.fit(X_train, y_train)
```

```python
accuracy_score(y_test, xgb_clf.predict(X_test))
```

```python
params = {
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

grid_search = GridSearchCV(xgb_clf, params, cv=3, scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
accuracy_score(y_test, grid_search.best_estimator_.predict(X_test))
```

```python
test_data = pd.read_csv('test.csv')
test_data.head()
```

```python
date_cols = ['date première tentative', 'date livraison/échec', 'date dernier statut']

for col in date_cols:
    test_data[col+' duration'] = (pd.to_datetime(test_data[col]) - pd.to_datetime(test_data['date expédition'])).dt.total_seconds() / (60*60*42)
```

```python
test_data['stop desk'] = test_data['stop desk?'].apply(lambda x: 0 if x=='oui' else 1)
```

```python
test_data['destination_commune_wilaya_same'] = (test_data['destination commune'] == test_data['destination wilaya']).astype(int)
```

```python
test_data = test_data.drop(columns=['Tracking', 'date expédition', 'date première tentative', 'date livraison/échec', 'date dernier statut', 'stop desk?', 'destination commune', 'depart commune'])
```

```python
test_data['ID expediteur'] = test_data['ID expediteur'].map(d)
```

```python
test_data.fillna(0, inplace=True)
```

```python
test_data.head()
```

```python
test_data_prepared = preprocessing.transform(test_data)
```

```python
test_data_prepared = test_data_prepared.toarray()
test_data_prepared
```

```python
y_pred_final = xgb_clf.predict(test_data_prepared)
```

```python
final_dict = {
    'Tracking': pd.read_csv('test.csv')['Tracking'],
    'dernier statut': encoder.inverse_transform(y_pred_final)
}
```

```python
df = pd.DataFrame(final_dict)
df.head()
```
